```python
def clean_cost_column(col):
    # Extract digits and decimal points only
    return col.str.replace(r'[^0-9.]', '', regex=True).astype(float)

df['Accommodation cost'] = clean_cost_column(df['Accommodation cost'])
df['Transportation cost'] = clean_cost_column(df['Transportation cost'])
```

```python
from sklearn.preprocessing  import OneHotEncoder

ohe=OneHotEncoder()

cat_encoded= ohe.fit_transform(df[['Traveler gender','Traveler nationality','Accommodation type','Transportation type']])
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler= MinMaxScaler()

num_scaled= scaler.fit_transform(df[['Duration (days)','Traveler age','Accommodation cost','Transportation cost']])
```

```python
from scipy.sparse import hstack, csr_matrix
num_sparse= csr_matrix(num_scaled)

combined_features= hstack([cat_encoded,num_sparse])
```

```python
combined_features
```

```python
from sklearn.metrics.pairwise import cosine_similarity

similarity= cosine_similarity(combined_features)
```
